```python
# Import Modules
import os
import csv
import pathlib
import pandas as pd
import matplotlib.pyplot as plt

#Set path for file and open the csv
csvpath = os.path.join('..', 'Chicago_Energy_Benchmarking_20240408.csv')

with open(csvpath) as csvfile:
    buildingenergydata = pd.read_csv(csvfile)
```

```python
#Filter to just the Year, Property Name, and weather Normalized Source EUI and Total GHG Emissions columns

filtered_bedata = buildingenergydata[['Data Year', 'Property Name', 'Weather Normalized Source EUI (kBtu/sq ft)', 'Total GHG Emissions (Metric Tons CO2e)']]

#Rename columns
new_names = {'Data Year': 'Year',
             'Weather Normalized Source EUI (kBtu/sq ft)': 'Median Weather-Normalized Source Energy Use Intensity (kBTU/sq. ft.)',
             'Total GHG Emissions (Metric Tons CO2e)': 'Total Greenhouse Gas Emissions (Metric Tons of Co2e)'}

filtered_bedata = filtered_bedata.rename(columns=new_names)
```

```python
# Create this table exclusively for several notable buildings
# Filter data for The Art Institute of Chicago exclusively for years 2016, 2018, 2020, and 2022
aic_data = filtered_bedata[filtered_bedata['Property Name'] == 'The Art Institute of Chicago']
aic_data = aic_data[aic_data['Year'].isin([2016, 2018, 2020, 2022])]

#Find change in both quantities ove rtime
aic_diff1 = aic_data['Median Weather-Normalized Source Energy Use Intensity (kBTU/sq. ft.)'].diff()
aic_diff2 = aic_data['Total Greenhouse Gas Emissions (Metric Tons of Co2e)'].diff()

#Add changes to tables
aic_data['Change in Source Energy Use'] = aic_diff1
aic_data['Change in Greenhouse Gas Emissions'] = aic_diff2

aic_data
```
